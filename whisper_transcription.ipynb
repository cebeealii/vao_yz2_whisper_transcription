```python
import subprocess

subprocess.run("pip install -U openai-whisper", shell=True)
subprocess.run("apt-get update", shell=True)
subprocess.run("apt-get install -y ffmpeg", shell=True)

print("Kurulum tamam.")
```

```python
import torch

print("CUDA var mı?:", torch.cuda.is_available())

if torch.cuda.is_available():
    print("GPU adı:", torch.cuda.get_device_name(0))
else:
    print("GPU aktif değil.")
```

```python
from google.colab import files

uploaded = files.upload()          # dosya seçme penceresi açılır
audio_path = next(iter(uploaded))  # yüklenen dosyanın adı
print("Yüklenen dosya:", audio_path)
```

```python
import whisper
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

print("Base model yükleniyor...")
model_base = whisper.load_model("base").to(device)

print("Model yüklendi.")
```

```python
print("Deşifre başlıyor...")

result_base = model_base.transcribe(
    audio_path,
    language="tr"
)

print("\n--- BASE MODEL ÇIKTISI ---\n")
print(result_base["text"])
```

```python
# Dosya adı
base_filename = "transcript_base.txt"

# Kaydetme
with open(base_filename, "w", encoding="utf-8") as f:
    f.write(result_base["text"])

print("Dosya kaydedildi:", base_filename)
```
